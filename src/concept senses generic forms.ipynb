```python
import csv
import inflect
```

```python
engine = inflect.engine()
```

```python
data = []
with open("../data/concept_senses.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        data.append(row)
```

```python
for entry in data:
    singular = entry['concept'].replace("_", " ")
    plural = engine.plural(singular)
    entry['singular'] = singular
    entry['plural'] = plural
```

```python
# write back to csv
with open("../data/concept_senses.csv", "w") as f:
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    writer.writeheader()
    writer.writerows(data)
```
